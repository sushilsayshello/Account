```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Define the path to your CSV file
file_path = '/Users/admin/Downloads/AppliedProject1/Attack2018.csv'

def load_data(file_path):
    """
    Load data from a CSV file and separate features and target variable.

    Args:
    file_path (str): Path to the CSV file.

    Returns:
    X (pd.DataFrame): Features dataset.
    y (pd.Series): Target variable.
    """
    try:
        data = pd.read_csv(file_path)

        # Check if the 'target' column exists
        if 'target' in data.columns:
            X = data.drop('target', axis=1)  # Features
            y = data['target']  # Target variable
        else:
            raise ValueError("Column 'target' not found in the data")

        return X, y

    except Exception as e:
        print(f"Error loading data: {e}")
        raise

def preliminary_analysis(X, y):
    """
    Perform preliminary analysis including data splitting, scaling, and model training.

    Args:
    X (pd.DataFrame): Features dataset.
    y (pd.Series): Target variable.

    Returns:
    None
    """
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Standardize features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train a Random Forest classifier
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train_scaled, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test_scaled)

    # Evaluate the model
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    print("\nConfusion Matrix:")
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(conf_matrix)

    # Display the confusion matrix as a table
    conf_matrix_df = pd.DataFrame(conf_matrix, index=['Actual Neg', 'Actual Pos'], columns=['Predicted Neg', 'Predicted Pos'])
    print("\nConfusion Matrix Table:")
    print(conf_matrix_df)

def main():
    """
    Main function to execute the preliminary analysis.
    """
    try:
        # Load the data
        X, y = load_data(file_path)

        # Perform preliminary analysis
        preliminary_analysis(X, y)

    except Exception as e:
        print(f"Failed to perform preliminary analysis: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the dataset
data = pd.read_csv('/Attack2018.csv')

# Display the first few rows of the dataset to verify it's loaded correctly
print(data.head())

# Display the column names
print("Column Names:", data.columns)

# Separate features and target variable
X = data.drop('Label', axis=1)  # Drop the 'Label' column from features
y = data['Label']  # Target variable is 'Label'

# Handle categorical data (if any), e.g., encoding
X = pd.get_dummies(X, drop_first=True)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict and evaluate K-Nearest Neighbors
y_pred_knn = knn.predict(X_test)
print("K-Nearest Neighbors Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Precision:", precision_score(y_test, y_pred_knn, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_knn, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred_knn, average='weighted'))

# Initialize and train Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)

# Predict and evaluate Naive Bayes
y_pred_nb = nb.predict(X_test)
print("\nNaive Bayes Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print("Precision:", precision_score(y_test, y_pred_nb, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_nb, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred_nb, average='weighted'))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define chunk size
chunk_size = 10000
chunks = pd.read_csv('/Attack2018.csv', chunksize=chunk_size)

# Initialize empty list to collect chunks
data_chunks = []

# Process each chunk
for chunk in chunks:
    # Append chunk to list
    data_chunks.append(chunk)

# Concatenate all chunks into a single DataFrame (optional, based on memory)
data = pd.concat(data_chunks, ignore_index=True)

# Display the first few rows of the dataset
print(data.head())

# Select a subset of features
selected_features = ['Dst Port', 'Protocol', 'Flow Duration', 'Tot Fwd Pkts', 'Tot Bwd Pkts', 'Label']
data = data[selected_features]

# Separate features and target variable
X = data.drop('Label', axis=1)
y = data['Label']

# Handle categorical data (if any), e.g., encoding
X = pd.get_dummies(X, drop_first=True)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict and evaluate K-Nearest Neighbors
y_pred_knn = knn.predict(X_test)
print("K-Nearest Neighbors Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Precision:", precision_score(y_test, y_pred_knn, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_knn, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred_knn, average='weighted'))

# Initialize and train Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)

# Predict and evaluate Naive Bayes
y_pred_nb = nb.predict(X_test)
print("\nNaive Bayes Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print("Precision:", precision_score(y_test, y_pred_nb, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_nb, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred_nb, average='weighted'))
```
